```python
import torch
```

### **1D Tensor**

```python
x = torch.FloatTensor([0.,1.,0.,8.])
x
```

```python
x.shape
```

### **3D Tensor**

```python
x = torch.FloatTensor([[[0.,1.], [1.,0.,]],
                       [[1.,1.], [0.,0.]]])
x
```

```python
x.shape
```

### **Addition**

```python
x = torch.FloatTensor([[0.,1.], [1.,0.]])
y = torch.FloatTensor([[1.,1.], [0.,0.]])
x+y
```

```python
x = torch.FloatTensor([[0.,1.],[1.,0.]])
y = torch.FloatTensor([3])
x+y
```

### **Multiplication**

```python
x = torch.FloatTensor([[0.,1.], [1.,0.]])
y = torch.FloatTensor([[1.,1.], [0.,0.]])
x.matmul(y) ##matrix mult
```

```python
x*y ##element-wise
```

```python
x.mul(y) ##element-wise
```

### **Mean**

```python
x = torch.FloatTensor([[2.,2.], [4.,4.]])
```

```python
print(x.mean())
print(x.mean(dim=0))
print(x.mean(dim=1))
```

### **Concatenate**

```python
x = torch.FloatTensor([[2.,2.], [4.,4.]])
y = torch.FloatTensor([[5.,5.], [10.,10.]])
```

```python
print(torch.cat([x,y], dim=0))
print(torch.cat([x,y], dim=1))
```

### **View**

```python
x = torch.FloatTensor([[[0.,1.], [1.,0.]],
                       [[1.,1.], [0.,0.]]])
x.shape
```

```python
x.view(-1,2).shape
```

```python
x.view(-1,1,2).shape
```

### **Squeeze**
- *squeeze*: reduce by 1D
- *unsqueeze*: increase by 1D

```python
x = torch.FloatTensor([[[0.], [1.], [2.]]])
x.shape
```

```python
torch.squeeze(x, dim=0).shape
```

```python
torch.squeeze(x, dim=2).shape
```

```python
torch.unsqueeze(x, dim=0).shape
```

```python
torch.unsqueeze(x, dim=2).shape
```
