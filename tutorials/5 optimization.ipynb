## **Gradient**

```python
import torch

a = torch.tensor([2.], requires_grad=True)
b = torch.tensor([6.], requires_grad=True)
```

```python
Q = 3*a**3 - b**2
Q # loss
```

```python
print(a.grad)
print(b.grad)
```

```python
Q.backward()
```

```python
print(a.grad)
print(b.grad)
```
